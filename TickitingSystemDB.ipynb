# [dbo].[addReservation]

```sql
/****** Object:  StoredProcedure [dbo].[addReservation]    Script Date: 8/21/2022 9:40:02 AM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE procedure [dbo].[addReservation]
(@n nvarchar(50),
@m nvarchar(50),
@c int,
@ev int)
as

begin
	
	Insert into Reservation (Name, Mail, Cost, EventId)
	values (@n, @m, @c, @ev)

	select ID FROM Reservation where Mail = @m and EventId= @ev;
	
end
GO
```

# [dbo].[altPrice]

```sql
/****** Object:  StoredProcedure [dbo].[altPrice]    Script Date: 8/21/2022 9:40:02 AM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create procedure [dbo].[altPrice]
(@price int,
@block int,
@event int,
@row int
)
as

begin

	update Price 
	Set Price = @price
	where EventId = @event and BlockId = @block and Idx = @row;
end
GO
```

# [dbo].[fetchSeats]

```sql
/****** Object:  StoredProcedure [dbo].[fetchSeats]    Script Date: 8/21/2022 9:40:02 AM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE procedure [dbo].[fetchSeats]
(@ev int)
as

begin

	Select t.BlockId, t.SeatX, t.SeatY 
	From Ticket t
	Join Reservation r
	On r.ID = t.ReservationId
	where r.EventId = @ev
end
GO
```
